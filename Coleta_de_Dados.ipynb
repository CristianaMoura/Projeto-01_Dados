```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cujas coluna
# distance é maior do que 2000,da tabela "flight_activity".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.year,
    fa.month,
    fa.flights_booked,
    fa.total_flights,
    fa.distance,
    fa.points_accumulated
  FROM
    flight_activity fa
  WHERE
    fa.distance > 2000

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()
```

```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cujas coluna
# distance é maior do que 2000 e a coluna month é igual a 9,da tabela "flight_activity".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.year,
    fa.month,
    fa.flights_booked,
    fa.total_flights,
    fa.distance,
    fa.points_accumulated
  FROM
    flight_activity fa
  WHERE
    fa.distance > 2000 AND fa.month = 9

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()
```

```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cujas coluna
# distance é maior do que 2000 e a coluna points_accumulated é menor que 100 igual
# a 9,da tabela "flight_activity".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.year,
    fa.month,
    fa.flights_booked,
    fa.total_flights,
    fa.distance,
    fa.points_accumulated
  FROM
    flight_activity fa
  WHERE
    fa.distance > 2000 OR fa.points_accumulated < 100

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()
```

```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cujas coluna
#    loyalty_card é igual a Star da tabela "flight_loyalty_history".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
   *
  FROM flight_loyalty_history flh
  WHERE
     flh.loyalty_card = "Star"

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()

```

```python
# Selecione as colunas: loyalty_number, country, city, gender, loyalty_card,
#salary da tabela "flight_loyalty_history".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
   flh.loyalty_number,
   flh.country,
   flh.city,
   flh.gender,
   flh.loyalty_card
  FROM flight_loyalty_history flh

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()

```

```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna gender
#     é igual a Female da tabela "flight_loyalty_history".

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
   flh.loyalty_number,
   flh.country,
   flh.city,
   flh.gender,
   flh.loyalty_card
  FROM
   flight_loyalty_history flh
  WHERE
   flh.gender = "Female"
"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()

```

```python
# Selecione as colunas: loyalty_number, month, year,distance, flights_booked e total_flights,
#  da tabela "flight_activity", e recupere somente as linhas cuja coluna "flights_booked"
#  é maior do que 10 e a coluna "flight_booked" é menor do que 12, da tabela flight_activity.

# consulta dos dados nos banco de dados
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.month,
    fa.year,
    fa.flights_booked,
    fa.total_flights
  FROM
    flight_activity fa
  WHERE
    fa.flights_booked > 10 AND fa.flights_booked < 12

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# confere se deu certo
df_atividade.head()
```
